## Outliers Detection

```python
# Get all numerical columns
num_cols = df.select_dtypes(include="number").columns.tolist()

# Plot settings
cols = 3  # number of plots per row
rows = (len(num_cols) + cols - 1) // cols

plt.figure(figsize=(cols * 5, rows * 4))

for i, col in enumerate(num_cols, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(data=df, x=col, color='skyblue')
    plt.title(col)
    plt.gca().set_facecolor('whitesmoke')  # light background

plt.tight_layout()
plt.show()
```

##### No Outliers Found

## Time Series Analysis

```python
# Daily Sales Trend
daily_sales = df.groupby('Date')['Total_Amount'].sum().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=daily_sales, x='Date', y='Total_Amount', marker='o')
plt.title('Daily Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
daily_sales
```

```python
# Create 'Month' column
df['Month'] = df['Date'].dt.to_period('M').astype(str)
```

```python
# Monthly Sales Trend
monthly_sales = df.groupby('Month')['Total_Amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Total_Amount', marker='o', color='green')

# Add Values above the points
for i in range(len(monthly_sales)):
    plt.text(i, monthly_sales['Total_Amount'][i] + 10,  # +10 to move label above the point
             str(round(monthly_sales['Total_Amount'][i])), 
             ha='center', fontsize=9)
    
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
monthly_sales
```

```python
# Extract Day of the Week (0 = Monday, 6 = Sunday)
df['Day_of_Week'] = df['Date'].dt.day_name()
```

```python
# Sales by Day of the Week
sales_by_day = df.groupby('Day_of_Week')['Total_Amount'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]).reset_index()

# Plot
plt.figure(figsize=(10, 5))
barplot = sns.barplot(data=sales_by_day, x='Day_of_Week', y='Total_Amount', hue='Day_of_Week', legend=False, palette='dark:skyblue')

# Add value labels
for index, row in sales_by_day.iterrows():
    barplot.text(index, row['Total_Amount'] + 10,   # +10 for space above bar
                 f"{int(row['Total_Amount'])}", 
                 color='black', ha="center", fontsize=9)

plt.title("Total Sales by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

## Customer and Product Analysis

```python
# Gender-wise Total Spending
gender_sales = df.groupby('Gender')['Total_Amount'].sum().sort_values(ascending=False)

# Bar Chart
plt.figure(figsize=(5,4))
plt.bar(gender_sales.index, gender_sales.values, color=['#ff9999','#66b3ff'])
plt.title("Total Amount Spent by Gender")
plt.ylabel("Total Amount")
plt.xlabel("Gender")
plt.tight_layout()
plt.show()
```

```python
gender_sales
```

```python
# Product Category Distribution by Gender
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Product_Category', hue='Gender', palette='Set2')
plt.title('Product Category Distribution by Gender')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.tight_layout()
plt.show()
```

```python
# Total Transaction by Day of Week

# Group by day and count unique transactions
txn_by_day = df.groupby('Day_of_Week')['Transaction_ID'].nunique().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]).reset_index()

# Plot
plt.figure(figsize=(10, 5))
bar = sns.barplot(data=txn_by_day, x='Day_of_Week', y='Transaction_ID', hue='Day_of_Week', legend=False, palette='Set3')

# Value Labels
for i in range(len(txn_by_day)):
    value = txn_by_day['Transaction_ID'][i]
    bar.text(i, value + 1, str(value), ha='center', fontsize=9)

plt.title("Total Transactions by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Number of Transactions")
plt.tight_layout()
plt.show()
```

```python
# Age Distribution of Customers

bins = [18, 25, 35, 45, 60, 100]
labels = ['<18', '18-30', '31-45', '46-60', '60+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

agegroup_counts = df['Age_Group'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
sns.barplot(x=agegroup_counts.index, y=agegroup_counts.values, hue=agegroup_counts, legend=False, palette='coolwarm')
plt.title("Number of Customers by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
agegroup_counts
```

```python
# Numeric Columns to analyze
num_cols = ['Age', 'Quantity', 'Price_per_Unit', 'Total_Amount']

plt.figure(figsize=(12, 7))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    ax = sns.histplot(df[col], kde=True, bins=30, color='dodgerblue')
    plt.title(f'Distribution of {col}', fontsize=12, color='black')
    plt.xlabel(col, fontsize=10)
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', linewidth=0.5)

plt.suptitle("Univariate Analysis of Numerical Features", fontsize=14, color='black', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Heatmap  Correlation Between Numerical Variables
plt.figure(figsize=(8, 5))
corr = df[['Age', 'Quantity', 'Price_per_Unit', 'Total_Amount']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```
